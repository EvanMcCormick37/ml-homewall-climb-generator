```python
from preprocess_climb_data import load_and_preprocess_climbs, augment_dataset
from generate_climbs import ClimbGenerator, ClimbMLP, ClimbGenerator, extract_hold_ids_from_holds, train_model, train_climb_model
import numpy as np
from typing import List, Tuple
```

```python
sequences, climbs, grades = load_and_preprocess_climbs("data/spraywall-climbs.json")
model, hold_features_map, train_idx = train_climb_model(num_epochs=1000)
```

```python
null_hold_idx=263
generator = ClimbGenerator(model, hold_features_map, null_hold_idx=null_hold_idx)
    
# Example: Use actual hold_ids from your wall
start_holds = [206, 139, 243, null_hold_idx]  # LH=206, RH=139, LF=243, RF=null
    
print(f"Generating climb from starting holds: {start_holds}")
print(f"(Null hold index: {null_hold_idx})")
    
sequence, embeddings = generator.generate_climb(
    start_holds=start_holds,
    max_moves=20,
    temperature=1.0
)
    
print(f"\nGenerated sequence:")
for i, holds in enumerate(sequence):
    display_holds = [h if h != null_hold_idx else -1 for h in holds]
    print(f"  Move {i}: LH={display_holds[0]:3d} RH={display_holds[1]:3d} LF={display_holds[2]:3d} RF={display_holds[3]:3d}")
```
