```python
import sqlite3
import pandas as pd
import os
import PIL
import json
from image_combine import center_overlay_images
```

```python
BOARDLIB = "data/boardlib"
KILTER = f"{BOARDLIB}/tension.db"
IMAGES_DIR = f"{BOARDLIB}/tension-images"

def get_layout_image_sets(database):
    with sqlite3.connect(database) as connection:
        results = connection.execute(
            """
            SELECT
                l.name layout_name,
                s.name product_size_name,
                p.layout_id,
                p.product_size_id,
                p.image_filename
            FROM product_sizes_layouts_sets p
            INNER JOIN 
                (SELECT id, name FROM layouts) AS l ON p.layout_id = l.id
            INNER JOIN
                (SELECT id, name FROM product_sizes) AS s ON p.product_size_id = s.id;
            """
        )
    
    image_dict = {}
    for row in results:
        title = f"{row[0]} {row[1]} {row[2], row[3]}"
        if title not in image_dict:
            image_dict[title] = []
        image_dict[title].append(row[4])
    
    return image_dict

image_sets = get_layout_image_sets(KILTER)
for output_name, image_names in image_sets.items():
    center_overlay_images(IMAGES_DIR,image_names, f"{IMAGES_DIR}/combined/{output_name}.png")
```
